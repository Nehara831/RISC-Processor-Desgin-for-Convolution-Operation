```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def apply_top_edge_filter(image):
    kernel = np.array([[-1, -1, -1],
                       [ 0,  0,  0],
                       [ 1,  1,  1]])

    image_height, image_width = image.shape

    output = np.zeros((image_height - 2, image_width - 2))

    for i in range(1, image_height - 1):
        for j in range(1, image_width - 1):
            sum = 0
            for ki in range(-1, 2):
                for kj in range(-1, 2):
                    sum += image[i + ki, j + kj] * kernel[ki + 1, kj + 1]
            output[i - 1, j - 1] = sum

    return output

def visualize_images(original, filtered):
    plt.figure(figsize=(10, 5))

    # Original image
    plt.subplot(1, 2, 1)
    plt.imshow(original, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # Filtered image
    plt.subplot(1, 2, 2)
    plt.imshow(filtered, cmap='gray')
    plt.title('Filtered Image')
    plt.axis('off')

    plt.show()

def load_and_resize_image(image_path):
    # Open the image file
    with Image.open(image_path) as img:
        # Convert the image to grayscale
        img_gray = img.convert('L')

        # Resize the image to 256x256
        img_resized = img_gray.resize((256, 256))

        # Convert the image to a NumPy array
        image_array = np.array(img_resized)

        return image_array

# Example usage
# Generate a random 256x256 image for demonstration
original_image = load_and_resize_image("test.jpeg")
original_image_rand = np.random.rand(256, 256)
print(original_image.shape)
print(original_image_rand.shape)
# Apply the top edge filter
filtered_image = apply_top_edge_filter(original_image)

# Visualize the original and filtered images
visualize_images(original_image, filtered_image)
```
